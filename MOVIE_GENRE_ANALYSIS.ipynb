```python
import pandas as pd
movies= pd.read_csv(r'C:\Users\Afzal\Downloads\imdb_movies.csv')
movies.describe(include= 'all')
```

```python
movies['Profit']= movies['revenue'] - movies['budget']
```

```python
movies.dropna(subset= ['genres'], inplace= True)
```

```python
movies.info()
```

```python
movies[movies.duplicated()]
```

```python
movies.drop_duplicates(inplace= True)
```

```python
movies_genres= movies[['popularity','budget','revenue','original_title','runtime','genres','release_year','vote_count','vote_average', 'Profit']]
movies_genres
```

```python
from pandas import Series, DataFrame
split = movies_genres['genres'].str.split('|').apply(Series,1).stack()
split.index= split.index.droplevel(-1)
split.name= 'genres_split'

movie_genre= movies_genres.join(split)
movie_genre
```

```python
del movie_genre['genres']
```

```python
movie_genre
```

Research Questions (Q):
1. Which genres are the most common (number of movies made)?


```python
count_genres= pd.DataFrame(movie_genre.groupby('genres_split').original_title.nunique()).sort_values('original_title', ascending= True)
count_genres
```

```python
count_genres['original_title'].plot.pie(title = 'Movie Genre', autopct= '%1.1f%%', figsize = (10,15))
```

```python
count_genres['original_title'].plot.barh(title = 'Movie Genre', figsize = (5,10), color= 'DarkBlue')
```

2. Which genres have high avg. budget and revenue?


```python
movie_genre
```

```python
avg_genres = movie_genre.groupby('genres_split').mean(numeric_only= True)
pd.options.display.float_format=  '{:2f}'.format
avg_genres
```

```python
avg_genres.sort_values('budget', ascending= True, inplace= True)
```

```python
avg_genres[['budget','revenue']].plot.barh(title = 'Budget and Revnue by genre', figsize = (5,10), color= ('DarkBlue','c'))
```

2.5 Which genres have high avg. Profit?

```python
avg_genres.sort_values('Profit', ascending= True, inplace= True)
avg_genres['Profit'].plot.barh(title = 'Profit by genre', x= 'genres_split'figsize = (5,10), color= ('c'))
```

3. Which genres have high avg. popularity?


```python
avg_genres.sort_values('popularity', ascending= True, inplace= True)
avg_genres['popularity'].plot.barh(title = 'popularity by genre', x= 'genres_split',figsize = (5,10), color= ('c'))
```

4. Which genres have highest number of movies with an voting avg. >=8?

```python
movie_genre
```

```python
movie_genre.describe()
```

```python
vote_fifty = movie_genre[(movie_genre['vote_count'] >= 250) & (movie_genre['vote_average'] >= 6)]
vote_zero = movie_genre[(movie_genre['vote_average'] >= 8)]
```

```python
vote_zero
```

```python
avg_genres= pd.DataFrame(vote_zero.groupby('genres_split').original_title.nunique()).sort_values('original_title', ascending= True)
avg_genres
```

```python
avg_genres.sort_values('original_title', ascending= True, inplace= True)
avg_genres['original_title'].plot.barh(title = 'Average vote by genre', figsize = (5,10), color= ('c'))
```

```python
avg_genres1= pd.DataFrame(vote_fifty.groupby('genres_split').original_title.nunique()).sort_values('original_title', ascending= True)
avg_genres1
```

```python
avg_genres1.sort_values('original_title', ascending= True, inplace= True)
avg_genres1['original_title'].plot.barh(title = 'Average vote by genre', figsize = (5,10), color= ('c'))
```
