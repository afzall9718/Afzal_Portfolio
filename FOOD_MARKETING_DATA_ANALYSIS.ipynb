Transformed categorical marital status across multiple columns into numerical values (1-5) 
based on a predefined mapping, sum these numerical values to create a single column representing the aggregated marital status.

```python
df['marital_Divorced']= df['marital_Divorced'].replace({1:5,0:0})
df['marital_Married']= df['marital_Married'].replace({1:4,0:0})
df['marital_Single']= df['marital_Single'].replace({1:3,0:0})
df['marital_Together']= df['marital_Together'].replace({1:2,0:0})
df['marital_Widow']= df['marital_Widow'].replace({1:1,0:0})
df['Marital_Status']= df[['marital_Divorced','marital_Married','marital_Single','marital_Together','marital_Widow']].sum(axis= 1)
```

```python
df['Marital_Status']=df['Marital_Status'].replace({5:'Divorced',4:'Married', 3:'Single', 2:'Together', 1:'Widow'})
df.head()
```

Transformed categorical education status across multiple columns into numerical values (1-5) 
based on a predefined mapping, sum these numerical values to create a single column representing the aggregated education level.

```python
df['education_2n Cycle']= df['education_2n Cycle'].replace({1:1,0:0})
df['education_Basic']= df['education_Basic'].replace({1:2,0:0})
df['education_Graduation']= df['education_Graduation'].replace({1:3,0:0})
df['education_Master']= df['education_Master'].replace({1:4,0:0})
df['education_PhD']= df['education_PhD'].replace({1:5,0:0})
df['Education_Status']= df[['education_2n Cycle','education_Basic','education_Graduation','education_Master','education_PhD']].sum(axis= 1)
```

```python
df['Education_Status']=df['Education_Status'].replace({5:'PHD',4:'Masters', 3:'Graduation', 2:'Basic',1: '2nd_Cycle'})
df.head()
```

A function has been created that assigns customers to age groups based on their respective ages

```python
df['Age'].sort_values()
age_groups = [(23,30),(31,40),(41,50),(51,60),(61,70),(71,85)]

def assign_age_group(Age):
    for age_range in age_groups:
        if age_range[0]<= Age  <= age_range[1]:
            return f"{age_range[0]} - {age_range[1]}"
    return("Unknown")
df['Age_Group']= df['Age'].apply(assign_age_group)
```

#Married, single, and together are the marital status categories that exhibit higher spending compared to other groups. To maximize revenue, the client should prioritize targeting and engaging these segments.

### **Overall Findings**

1. **Age Group Analysis**:
   - Customers aged between 30 and 70 tend to spend more money overall. However, they are less likely to respond positively to marketing campaigns. Despite their lower campaign acceptance rate, this age group represents a significant portion of the customer base, making them a key demographic for revenue generation.

2. **Campaign Channel Preferences**:
   - Customers who shop through catalogs are more likely to accept marketing campaigns, while those who shop in person tend to spend more money. To optimize marketing efforts, it is recommended to allocate resources across all channels with the following distribution:
     - **40% to catalog campaigns**
     - **30% to in-store promotions**
     - **30% to web-based campaigns**

3. **Family Structure**:
   - Customers with fewer or no children tend to spend more. Therefore, marketing efforts should prioritize targeting individuals or couples without kids to maximize returns.

4. **Education Level**:
   - Education level does not significantly impact spending behavior or campaign acceptance. As a result, it is not necessary to tailor marketing strategies based on educational background.

5. **Marital Status**:
   - Marital status has a minimal impact on spending patterns. However, individuals who are married, single, or living together tend to spend more compared to other marital categories. These groups should be considered in marketing strategies.

---

### **Key Revenue-Generating Segments**

1. **Primary Target Audience**:
   - Middle-aged individuals (30-70 years old) with higher income levels and no children are the most profitable segment. To effectively reach this group, marketing efforts should be distributed across multiple platforms using the recommended channel split (40% catalog, 30% in-store, 30% web).

---

### **Opportunities for New Customer Acquisition**

2. **Focus on Younger and Older Age Groups**:
   - Younger customers (21-30 years old) and older customers (70+ years old) have shown a higher likelihood of accepting marketing campaigns. While their spending may not be as high as the middle-aged group, their responsiveness to campaigns makes them a valuable segment for driving engagement and acquiring new users.
